### Dropout,earlystopping

```python
import keras 
from keras.models import Sequential

from keras.layers import InputLayer,Dense,Dropout
from sklearn.datasets import make_classification
```

```python
fv,cv=make_classification(n_samples=10000,n_features=2,n_informative=2,n_redundant=0,n_repeated=0,n_classes=2,class_sep=3,random_state=10)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.scatterplot(x=fv[:,0],y=fv[:,1],hue=cv)
plt.show
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(fv,cv,test_size=0.2,random_state=10,stratify=cv)
```

```python
from sklearn.preprocessing import StandardScaler
std=StandardScaler()
```

```python
x_traint=std.fit_transform(x_train)
x_testt=std.fit(x_test)
```

```python
model=Sequential()
model.add(InputLayer(input_shape=(2,)))
model.add(Dense(units=34,activation="tanh",use_bias=True,))
model.add(Dense(units=23,activation="tanh",use_bias=True,))
model.add(Dense(units=18,activation="tanh",use_bias=True,))
model.add(Dense(units=16,activation="tanh",use_bias=True,))

model.add(Dense(units=1,activation="sigmoid",use_bias=True,))
```

```python
model.summary()
```

```python
model.compile(optimizer="sgd",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
hist=model.fit(x_traint,y_train,batch_size=30,epochs=100,validation_split=0.2)
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1,101),hist.history["loss"],label="training loss")
plt.plot(range(1,101),hist.history["val_loss"],label="val loss")
plt.legend()
plt.show()
```

```python
model=Sequential()
model.add(InputLayer(input_shape=(2,)))
model.add(Dense(units=34,activation="tanh",use_bias=True,))
model.add(Dropout(rate=0.3))
model.add(Dense(units=23,activation="tanh",use_bias=True,))
model.add(Dropout(rate=0.4))
model.add(Dense(units=18,activation="tanh",use_bias=True,))
model.add(Dropout(rate=0.5))
model.add(Dense(units=16,activation="tanh",use_bias=True,))
model.add(Dropout(rate=0.5))

model.add(Dense(units=1,activation="sigmoid",use_bias=True,))
```

```python
model.summary()
```

```python
from keras.callbacks import EarlyStopping
```

```python
es=EarlyStopping(monitor='val_loss',
    min_delta=0.01,
    patience=2,
    verbose=1,
    mode='auto',
    baseline=None,
    restore_best_weights=False,
    start_from_epoch=10,)
```

```python
model.compile(optimizer="sgd",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
hist=model.fit(x_traint,y_train,batch_size=30,epochs=1000,validation_split=0.2,callbacks=[es])
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1,14),hist.history["loss"],label="training loss")
plt.plot(range(1,14),hist.history["val_loss"],label="val loss")
plt.legend()
plt.show()
```

```python
def f1(hp):
    model=Sequential()
    model.add(InputLayer(input_shape=(5,)))
    co=0
    for layer in range(hp.Int("layers",min_value=3,max_value=100)):
        model.add(Dense(units=hp.Int("no of neurons {}".format(co),min_value=3,max_value=100),activation=hp.Choice("1hl activation",values=["sigmoid","tanh","relu"])))
        model.add(BatchNormalization())
        co+=1
    model.add(Dense(units=4,activation="softmax"))
    model.compile(optimizer=hp.Choice("optimizer",values=["sgd","adagrad","adadelta","adam"]),loss="sparse_categorical_crossentropy",metrics=["accuracy"])
    return model
```

```python
hb=kt.Hyperband(f1,objective="val_loss",max_epochs=10,factor=3,directory="k1") #factor is saying how many model you want to remove
```

```python
hb.search(fv,cv,validation_split=0.1,epochs=5)
```
